# QA-ассистент

## Цель

Построить QA-ассистента отеля на LangGraph, прогнать его на эталонном наборе вопросов и оценить качество ответов.

## Входные данные

- `hotel_booklet.docx` — буклет с информацией об отеле (источник знаний для ассистента).
- `hotel_qa.json` — 20 эталонных пар «вопрос → ответ», самые частые запросы гостей. Используется **только** для оценки.

## Что нужно сделать

1. **Ассистент.** Реализовать на LangGraph граф, который принимает вопрос гостя и возвращает ответ, опираясь на содержимое буклета.
   - Загрузите буклет один раз и передавайте его ассистенту как контекст (или через RAG, если решите так).
   - **Важно:** ответы из `hotel_qa.json` нельзя передавать ни в промпт, ни в память ассистента. Они нужны только как разметка.

2. **Оценка.** Прогнать ассистента по всем 20 вопросам из json и сравнить ответы с эталонными.
   - Выберите **один** способ оценки сходства текстов (ANLS, LLM-as-a-Judge, косинусная близость эмбеддингов и т.п.).
   - **Обоснуйте**, почему выбрали именно этот метод (плюсы, минусы, почему подходит для данной задачи).

3. **Граф-пайплайн.** Сборка должна быть единым LangGraph: загрузка вопросов → ответ ассистента → вычисление метрики → агрегация → печать итогов.

## Что должно быть в результате

- Ноутбук с собранным и выполненным графом.
- Обоснование выбранной метрики.
- Финальная сводка по 20 вопросам: метрика по каждому + агрегированный показатель (например, среднее).

## Бонус

- Сравнить **две** метрики оценки и обсудить, в каких случаях они расходятся.
- Добавить узел-классификатор «вопрос относится к отелю / нет» и обработку out-of-scope запросов.
